```python
import pandas as pd
```

```python
rep = pd.read_csv("/outs/clean_rep.csv")
```

```python
rep.head()
```

```python
d = []
for row in rep.itertuples():
    for i in range(1000):
        new_row = row._replace(pos = row.pos-500+i)
        d.append(new_row)

d = pd.DataFrame(d)
d.head()
```

```python
d = d.drop("Index", axis=1)
d.head()
```
